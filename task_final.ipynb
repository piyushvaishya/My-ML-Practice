```python
frame = assign_data.groupby('Registration Number')["Assignment Number"].nunique().to_frame().reset_index()
```

```python
frame = frame.rename(columns= {0: 'list'})
```

```python
frame.sort_values("Assignment Number",ascending = False).head()
```

```python
final = pd.merge(git_data, frame, how='inner', left_on = 'Registration No.', right_on = 'Registration Number')
```

```python
final=final.drop(columns=['S.No.','Registration No.','Git Repo Link',])
final.sample(5)
```

```python
final = final.sort_values(by=['Assignment Number'], ascending=False)
```

```python
plt.figure(figsize=(15, 10))
gs=gridspec.GridSpec(4,6)
gs.update(wspace=0.0, hspace=0.0)


plt.subplot(gs[0,0])
plt.xticks(())
plt.yticks(())
plt.text(0.5, 0.5, 
         'Logo\nComing\n  Soon', 
         ha='center', va='center',
         fontsize=20, 
         color="black")

plt.subplot(gs[1,0])
plt.xticks(())
plt.yticks(())
plt.text(0.5, 0.5, 
         "Stars Earned\n\n  {}".format(final['Stars ⭐️'].sum()), 
         ha='center', va='center',
         fontsize=18, 
         color="black")

plt.subplot(gs[2:3:4,0])
plt.xticks(())
plt.yticks(())
plt.text(0.5, 0.5, "Total\nSubmissions\n\n {}".format(final['Assignment Number'].sum()), ha='center', 
         va='center',fontsize=18, color="black")

plt.subplot(gs[3:,:1])
plt.xticks(())
plt.yticks(())
plt.text(0.5, 0.5, "Stars\nEarned\n\n  {}".format(int(final['Stars ⭐️'].sum())), ha='center', 
         va='center',fontsize=18, color="black")

plt.subplot(gs[:,1:2])
plt.xticks(())
plt.yticks(())

ax6=plt.subplot(gs[:,2:4])
ax6.set_facecolor('whitesmoke')
plt.barh(length,stars['Stars ⭐️'],color='gold')
plt.xticks(())
plt.title('Star Performer')
for i in range(len(stars['Name'])):
    plt.text(stars['Stars ⭐️'].values[i]-0.5 ,i,int(stars['Stars ⭐️'][i]), size = 12)

colors =['palevioletred','coral','sandybrown','lightgreen','yellowgreen','mediumseagreen','mediumturquoise']
ax7=plt.subplot(gs[:,4:])
plt.xticks(())
plt.yticks(())
ax7=ax7.table(cellText=final[['Name','Assignment Number']].values,colLabels=['Names','Last \n Assignment'],
              colWidths = [0.6,0.4],cellLoc = 'left',loc='center')
ax7.auto_set_font_size(False)
ax7.set_fontsize(11)
for i in range((final['Name'].count()+1)):
    if i==0:
        ax7[(i,0)].set_text_props(color='w')
        ax7[(i,1)].set_text_props(color='w')
        ax7[(i,0)].set_facecolor(color='dimgray')
        ax7[(i,1)].set_facecolor(color='dimgray')
    else:
        ax7[(i,0)].set_facecolor(color=colors[final['Assignment Number'].values[i-1]])
        ax7[(i,1)].set_facecolor(color=colors[final['Assignment Number'].values[i-1]])
ax7.scale(1,1.7)
```
