```python
#importing lib
import numpy as np
import pandas as pd
```

```python
#importing dataset
dataset = pd.read_csv("Assignment5.csv" ,names = ['employee', 'works_under'])
```

```python
dataset
```

```python
result = []
for i in range(dataset.employee.count()):
    temp = []
    temp.append(dataset.employee[i])
    for j in range(i,dataset.employee.count()):
        if dataset.employee[j]!=dataset.works_under[j]:
            temp.append(dataset.works_under[j])
        else:
            break
    #print(temp)
    result.append(temp)
```

```python
hierarchy = pd.DataFrame(result)
```
